Cleaning Data

replacing null with average

```python
columns_to_replace = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']

print(air_polution_df[columns_to_replace].dtypes)

for column in columns_to_replace:
    column_mean = air_polution_df[column].mean()
    air_polution_df[column].fillna(column_mean, inplace=True)
```

Pertanyaan 1:
bagaimana kondisi PM10 setiap tahun
measuring PM 10 setiap tahun

```python
plt.figure(figsize=(15,6))
sns.lineplot(x='year', y='PM10', data=air_polution_df)
plt.title('PM10 over the year')
plt.xlabel('year')
plt.ylabel('PM10 polutant')

```

Pertanyaan 2:
bagaimana kondisi polutan yang selalu melebihi standards setiap tahun.
mengukur konsentrasi polutan setiap tahun untuk melihat polutan mana yang melebihi standard di setiap tahun

```python
pollutants = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
fig, axes = plt.subplots(nrows=len(pollutants), figsize=(12, 6 * len(pollutants)))

for i, pollutant in enumerate(pollutants):
    max_concentration_per_year = air_polution_df.groupby(['year', 'station'])[pollutant].max().reset_index()
    heatmap_data = max_concentration_per_year.pivot_table(index='year', columns='station', values=pollutant, aggfunc='max')

    sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=.5, ax=axes[i])
    axes[i].set_title(f'Station with the Highest {pollutant} Concentration Each Year')

plt.tight_layout()
plt.show()
```

from those visualiasation, we know that every year, has a various pollutant. but inst comparable because every pollutant has a different maximum number in air. well its no  means this pollutant more than other is bad.
